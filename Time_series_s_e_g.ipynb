```python
import pandas as pd

df_earnings_final = pd.read_csv("/content/earnings (3).csv")
import googletrans

from googletrans import Translator
translator = Translator()
month = 'Maio'

print(translator.translate(month,src='pt', dest='en').text)

df_earnings_final_months = df_earnings_final['Date'].str.split('de', expand=True)
df_earnings_final_months.rename(columns = ({0: "month", 1: "year"}),inplace=True)
df_earnings_final_months['month'] = df_earnings_final_months['month'].apply(lambda month: translator.translate(month,src='pt', dest='en').text)
```

```python
df_earnings_final
```

```python
df_earnings_final_data = df_earnings_final.drop(columns=['Unnamed: 0'])
```

```python
df_earnings_final_data
```

```python
df_earnings = df_earnings_v1.drop(columns=['Date'])
```
