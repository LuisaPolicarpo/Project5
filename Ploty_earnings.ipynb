```python
import pandas as pd
```

```python
df_e_2022= pd.read_csv('https://raw.githubusercontent.com/LuisaPolicarpo/Project5/main/earnings_2022.csv', encoding='latin-1', sep=';')
```

```python
df_e_2022
```

```python
df_2022_e_e = df_e_2022.rename(columns = {'Período de referência dos dados':'Date','Tipo':'Type', 'Proveitos de aposento': 'Value'})
```

```python
df = df_2022_e_e.drop(columns='Unnamed: 3')
```

```python
df_ear_2017 = pd.read_csv('https://raw.githubusercontent.com/LuisaPolicarpo/Project5/main/Dataset/earnings.csv')
```

```python
df_ear_2017
```

```python
df_final_earnings= pd.concat([df, df_ear_2017], axis=0)
```

```python
df_final_earnings['Date'] = df_final_earnings['Date'].str.replace(' de ', '/')
df_final_earnings[['Month', 'Year']] = df_final_earnings['Date'].str.split('/',1,expand=True)
```

```python
months = {'Janeiro':'01', 'Fevereiro':'02', 'Março':'03', 'Abril':'04', 'Maio':'05', 'Junho':'06', 'Julho':'07', 'Agosto':'08', 'Setembro':'09', 'Outubro':10, 'Novembro':11, 'Dezembro':12}

df2=df_final_earnings.replace({"Month": months}) #convert the name to number of month
df2['Year'] = df2['Year'].astype(str)
df2['Month'] = df2['Month'].astype(str)
# df2['Date'] = df2['Month']+df2['Year']

df_22 = df2[['Date', 'Type', 'Value']]

# # df_22['Date'] = pd.to_datetime(df_22['Date'], format='%m%Y')
# df_22 = df_22.sort_values('Date')
# df_final_earnings_2 = df_22.set_index('Date')
```
